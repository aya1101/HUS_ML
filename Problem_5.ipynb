```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error
```

**Fit a linear regression model to the data and report the estimated parameters.**

```python
np.random.seed(42)  # For reproducibility
n_samples = 100
x = np.random.uniform(0, 10, n_samples)
epsilon = np.random.normal(0, np.sqrt(4), n_samples)
y = 2 * x + epsilon
x = x.reshape(-1,1)
```

```python
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse}")
```

**Fit a 9th-degree polynomial regression model to the same data.**

```python
poly_model = make_pipeline(PolynomialFeatures(9), LinearRegression())
poly_model.fit(x, y)
y_pred_poly = poly_model.predict(x)
poly_mse = mean_squared_error(y, y_pred_poly)
```
